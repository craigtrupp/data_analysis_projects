```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('popular-languages.csv', index_col=0)
df.head()
```

```python
df.columns
```

```python
df['AverageAnnualFloat'] = df['Average Annual Salary']
df['AverageAnnualFloat'] = df['AverageAnnualFloat'].apply(lambda x: x.replace('$', ''))
df['AverageAnnualFloat'] = df['AverageAnnualFloat'].apply(lambda x: x.replace(',', ''))
df['AverageAnnualFloat'] = df['AverageAnnualFloat'].astype('float')
print(df.head(), '\n')

df_values_sorted = df.sort_values(by='AverageAnnualFloat', ascending=False).reset_index()
df_values_sorted
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
sns.set_theme(style="whitegrid", palette="pastel")
sns.barplot(x=df_values_sorted['Language'], y=df_values_sorted['AverageAnnualFloat'])
ax.bar_label(ax.containers[0])
plt.ylim(df_values_sorted['AverageAnnualFloat'].min() - 2500,df_values_sorted['AverageAnnualFloat'].max() + 2500)
plt.xticks(rotation=35)
plt.ylabel('Average Annual Salary ($)')
plt.show()
```

```python
job_postings = pd.DataFrame({
    'Location' : ['New York', 'Washington D.C.', 'San Francisco', 'Austin', 'Los Angeles', 'Seattle', 'Detroit'],
    'JobCount': [3226, 5316, 435, 435, 640, 3375, 3945]

})
job_postings = job_postings.sort_values('JobCount', ascending=False).reset_index(drop=True)
job_postings
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(2, 1, figsize=(10,12))
# ax[0].bar(x=job_postings['Location'], height=job_postings['JobCount'])
# ax[1].bar(x=job_postings['Location'], height=job_postings['JobCount'])
colors = ['lightcoral', 'lightsteelblue', 'forestgreen', 'slateblue', 'blueviolet', 'tomato', 'darkorange']
top = ax[0].bar(x=job_postings['Location'], height=job_postings['JobCount'], 
                color = colors)
bottom = ax[1].barh(y=job_postings['Location'], width=job_postings['JobCount'],
                   color = colors)
ax[0].bar_label(top)
ax[0].set_ylabel('Total Job Counts For City')
ax[1].bar_label(bottom)
ax[1].set_xlabel('Total Job Counts For City')
fig.suptitle('Job Counts for Cities from Job API Returns', fontsize=30)
plt.show()
```
