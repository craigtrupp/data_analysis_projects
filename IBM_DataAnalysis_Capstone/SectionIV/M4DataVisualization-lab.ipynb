## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
query = """
SELECT * 
FROM DatabaseDesireNextYear
"""
dframe_dbdesired = pd.read_sql_query(query, conn)
print(dframe_dbdesired.head())
print(dframe_dbdesired.shape)
```

```python
query = """
SELECT DatabaseDesireNextYear as DbDesired, COUNT(*) as RespondentCount
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY RespondentCount DESC
LIMIT 5
"""
dbase_dframe_top5 = pd.read_sql_query(query, conn)
print(dbase_dframe_top5)
```

```python
dbase_dframe_top5 = dbase_dframe_top5.assign(
    Percentage = lambda x: round(x['RespondentCount'] * 100 / dbase_dframe_top5['RespondentCount'].sum(), 2))
display(dbase_dframe_top5)
```

```python
explode_list = [0.0, 0, 0.01, 0.02, 0.03]
dbase_dframe_top5['RespondentCount'].plot(kind='pie',
                        figsize=(13, 6),
                        autopct='%1.1f%%', 
                        startangle=190,    
                        shadow=True,       
                        labels=None,                 # turn off labels on pie chart
                        pctdistance=0.75            # the ratio between the pie center and start of text label
                        #explode=explode_list
                        )

plt.title('Respondent Top Desired DB to Learn Next Year', y=1.08) 
plt.axis('equal')
plt.legend(dbase_dframe_top5['DbDesired'], loc='upper left')
plt.show()
```

#### Question 3 : How many respondents indicated that they currently work with 'SQL'? 
#### Question 4: How many respondents indicated that they work on 'MySQL' only?

```python
query_db_worked_with = """
SELECT * 
FROM DatabaseWorkedWith
"""
dframe_dbworkwith = pd.read_sql_query(query_db_worked_with, conn)
display(dframe_dbworkwith.head())
print(dframe_dbworkwith['DatabaseWorkedWith'].value_counts(ascending=False))
```

```python
print(len(dframe_dbworkwith.query('DatabaseWorkedWith.str.contains("SQL")', engine='python')))
print(len(dframe_dbworkwith))
```

```python
sql_only_query = """
SELECT Respondent, COUNT(DatabaseWorkedWith) AS TotalDBWorkedWith
FROM DatabaseWorkedWith
GROUP BY Respondent
HAVING COUNT(DatabaseWorkedWith) = 1 AND DatabaseWorkedWith='MySQL'
"""
sql_only_dframe = pd.read_sql_query(sql_only_query, conn)
display(sql_only_dframe)
```

#### The Db of interest (DatabaseWorkedWith) has a 1 to n type relationship with the respondent index value and database worked with. Likely need a groupby to drill down into individual numbers as the question allude to

```python
sql_respondents_dframe = dframe_dbworkwith.query('DatabaseWorkedWith.str.contains("SQL")', engine='python')
```

```python
display(sql_respondents_dframe.head())
display(sql_respondents_dframe['DatabaseWorkedWith'].value_counts())
```

```python
## Unstack a grouped by count 
agg_respondendt_db_worked_with_frame = sql_respondents_dframe.groupby('Respondent').value_counts().unstack().fillna(0)
display(agg_respondendt_db_worked_with_frame.head())
print([(agg_respondendt_db_worked_with_frame[x].sum(), x) for x in agg_respondendt_db_worked_with_frame.columns.tolist()])
## Below is the unstacked (wide) dataframe for respondents showing all the db they worked with
## Also is a quick list comprehension showing how many total counts there are for each respective service
```

```python
## These are the similar above counts we have but we can use the widened dataframe to find respondents who only
## have used MySQL and no other dataframe with the dataframe we have now
agg_small_copy = agg_respondendt_db_worked_with_frame.copy().iloc[:5, :]
display(agg_small_copy)
agg_small_copy['Total'] = agg_small_copy.sum(axis=1)
display(agg_small_copy)
```

```python
## Add Total row after confirming snippet for smaller dataframe worked
agg_respondendt_db_worked_with_frame['Total'] = agg_respondendt_db_worked_with_frame.sum(axis=1)
display(agg_respondendt_db_worked_with_frame.head(15))

## Now let's see if we can just get MySQL totals
mySQL = agg_respondendt_db_worked_with_frame['MySQL'] == 1
Total = agg_respondendt_db_worked_with_frame['Total'] == 1

## Conditional Setting for only MySQL usage in db worked with

print("The DB respondents who only solely worked with mySQL is : {}".format(
    len(agg_respondendt_db_worked_with_frame.loc[(mySQL) & (Total)])))

print('\n\n')
print(agg_respondendt_db_worked_with_frame.shape)
```

```python
mysql_only_dframe = agg_respondendt_db_worked_with_frame.loc[(mySQL) & (Total)]
len(mysql_only_dframe)
display(mysql_only_dframe)
```

```python
## Question 3 : How many respondents indicated that they currently work with 'SQL' (LanguageWorkedWith)
language_query = """
SELECT LanguageWorkedWith, COUNT(*) AS RespondentCountWorkedWith
FROM LanguageWorkedWith
GROUP BY LanguageWorkedWith
ORDER BY RespondentCountWorkedWith DESC
"""
langauge_worked_with = pd.read_sql_query(language_query, conn)
display(langauge_worked_with)
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
#display(dframe_master[['WorkWeekHrs', 'CodeRevHrs', 'Age']].head())
stacked_dframe = dframe_master[['WorkWeekHrs', 'CodeRevHrs', 'Age']]
stacked_dframe_30_35 = stacked_dframe.loc[(stacked_dframe['Age'] >= 30) & (stacked_dframe['Age'] <= 35)]
```

```python
print(len(stacked_dframe),len(stacked_dframe_30_35))
display(stacked_dframe_30_35['Age'].value_counts())
```

```python
fig, ax = plt.subplots()
ax.bar('CodeRev & WorkWeek', stacked_dframe_30_35['WorkWeekHrs'].median(), width=0.15, label='WorkWeekHrs Median (30-35)')
ax.bar('CodeRev & WorkWeek', stacked_dframe_30_35['CodeRevHrs'].median(), width=0.15, 
       label='CodeRevHrs Median (30-35)', bottom=stacked_dframe_30_35['WorkWeekHrs'].median())
ax.set_ylabel('Hours')
plt.title('Median Hours for (30-35) Work Week and Code Review', y=1.05)
plt.legend()
plt.ylim(0,55)
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
#display(dframe_master.columns.tolist())
#display(dframe_master[['ConvertedComp', 'Age']].head())

## Steps 
# 1) Subset Dataframe to 45-60 Age Rows only
# 2) Groupby Age and Aggregate Median Compy count for each GroupedBy Age

dframe_45_60 = dframe_master[['Age', 'ConvertedComp']].loc[(dframe_master['Age'] >= 45) & 
                                                           (dframe_master['Age'] <= 60)]
print(len(dframe_45_60)) # length of subset dataframe
print('\n')
print(dframe_45_60['Age'].value_counts()) # counts of rows for categorical age binning

dframe_45_60_comp_agg = dframe_45_60.groupby('Age')['ConvertedComp'].agg('median')
print('\n')
display(dframe_45_60_comp_agg)
```

```python
dframe_45_60_comp_agg.plot(kind='line')
```

```python
#[dframe_45_60_comp_agg.index.tolist(), dframe_45_60_comp_agg.values.tolist()]
sns.lineplot(data=dframe_45_60_comp_agg, x=dframe_45_60_comp_agg.index, y=dframe_45_60_comp_agg.values)
plt.xlabel('Age')
plt.ylabel('Median Compensation : $')
plt.title('Median Compensation for Workers Aged (45-60)')
plt.show()
```
